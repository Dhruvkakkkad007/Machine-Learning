## Dataset Splitting with `train_test_split`

### Why do we split our dataset into Training and Testing sets?**
**Answer:**
If we train a model on our entire dataset, we cannot verify how it performs on new, unseen data. It might just memorize the training dataset (a problem known as **Overfitting**).


### How does `train_test_split` work? Show the code for splitting `iphone.csv` with a test size of 0.33.**
We separate our features ($X$, inputs) and targets ($y$, output/label) first, then split them using a test size of 0.33.

```python
x_iphone = df_iphone[['number']]
y_iphone = df_iphone['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_iphone,y_iphone,test_size=0.2,random_state=42)
```

### Show exactly what `X_train`, `X_test`, `y_train`, and `y_test` look like. Explain why specific rows ended up in which set.**
Run the following cell to inspect the resulting training and testing divisions.

```python
print("x_train shape:", x_train.shape);
print(x_train)
print("y_train shape:", y_train.shape);
print(y_train)


print("y_test shape:", y_test.shape);
print(y_test)

print("x_test shape:", x_test.shape);
print(x_test)
```

# Import LinearRegression model from linear_model family 

```python
from  sklearn.linear_model import LinearRegression

model = LinearRegression()
```

# Fit the data

```python
model.fit(x_train,y_train)
```

 # Predict price of iphone 18

```python
import warnings  
warnings.filterwarnings('ignore')

y_pred = model.predict([[18]])
print("Predicted price for iPhone 18:", y_pred)
```

 # Predict price of iphone 20 to 80

```python
import numpy as np

IphoneNumber = np.arange(20,80).reshape(-1,1)
ans = model.predict(IphoneNumber)
ans
```

```python
import matplotlib.pyplot as plt

plt.plot(IphoneNumber, ans, color='blue', label='Regression Line')
plt.xlabel('iPhone Number')
plt.ylabel('Predicted Price')
plt.title('Linear Regression: iPhone Price Prediction')
plt.legend()
plt.show()
```
